#### Data visualizzation

```python
plt.figure(figsize=(8,6))
sns.barplot(x='Sex',y='Survived',data=df,color = 'blue',edgecolor='black')
plt.title("Gender vs Surival Rate")
plt.xlabel("Gender")
plt.ylabel("Survived")
plt.show()
```

```python
#age distribution

plt.figure(figsize=(8,6))
sns.histplot(df['Age'],bins =20,color="teal",kde=True)
plt.show()
```

```python
sns.scatterplot(x='Age',y='Fare',data=df,hue=df['Pclass'],palette ='viridis')
plt.title('Relationship Fare and Age')
plt.show()
```

```python
sns.lineplot(x='PassengerId',y='Fare',data=df,marker='o',linestyle='-.')
plt.show()
```
