```python
#Distribution App Ratings:
plt.figure(figsize=(10,6))
sns.histplot(data['Rating'].dropna(),kde=True,bins=20,color='skyblue')
plt.title('Distribution of App Rating')
plt.xlabel('Rating')
plt.ylabel('Frequency')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=data,color='lightgreen')
plt.title('Bosplot of App Rating')
plt.xlabel('Rating')
plt.show()
```

```python
data['Rating']=np.log1p(data['Rating'])
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=data,color='lightgreen')
plt.title('Bosplot of App Rating')
plt.xlabel('Rating')
plt.show()
```

```python
Q1=data['Rating'].quantile(0.25)
Q3=data['Rating'].quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
```

```python
filterdata=data[(data['Rating']>=lower_bound)&(data['Rating']<=upper_bound)]
```

```python
plt.figure(figsize=(10,6))
sns.boxplot(x='Rating',data=filterdata,color='lightgreen')
plt.title('Boxplot of App Rating')
plt.xlabel('Rating')
plt.show()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(y='Category',data=data,order=data['Category'].value_counts().index,palette='viridis')
plt.title('count of apps by category')
plt.xlabel('count')
plt.ylabel('category')
```
